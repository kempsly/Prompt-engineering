```python
# import Lenny Newsletter - Sheet1.csv
import pandas as pd

df = pd.read_csv("Lenny Newsletter - Sheet1.csv")
df.head()
```

```python
# urls = list of "Title" and "Link" as a list of tuples
urls = list(zip(df["Title"], df["Link"]))

print(urls[:5])
```

```python
import os
import time
import requests
from bs4 import BeautifulSoup

for url in urls:
    try:
        link = url[1]
        filename = f"lenny/{link.split('/')[-1]}.txt"

        # Fetch the content from the URL

        response = requests.get(link)
        response.raise_for_status()  # Will raise an HTTPError if the HTTP request returned an unsuccessful status code

        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text()

        # Save the content to a file in a folder
        with open(f"lenny/{link.split('/')[-1]}.txt", "w") as file:
            file.write(text)

        # sleep for 3 seconds to avoid rate limiting
        time.sleep(3)

    except requests.RequestException as e:
        print(f"Error fetching {url}: {e}")
```

```python
import os
import time
import requests
from bs4 import BeautifulSoup

# Create the directory if it does not exist
directory = "lenny"
if not os.path.exists(directory):
    os.makedirs(directory)



for url in urls:
    try:
        link = url[1]
        filename = f"{directory}/{link.split('/')[-1]}.txt"

        # Fetch the content from the URL
        response = requests.get(link)
        response.raise_for_status()  # Raises an HTTPError for bad responses

        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text()

        # Save the content to a file in the folder
        with open(filename, "w", encoding='utf-8') as file:  # Using utf-8 encoding
            file.write(text)

        print(f"Successfully saved content from {link} to {filename}")

        # Sleep for 3 seconds to avoid rate limiting
        time.sleep(3)

    except requests.RequestException as e:
        print(f"Error fetching {link}: {e}")
    except Exception as e:
        print(f"An error occurred while processing {link}: {e}")
```
