```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

# Initialize the model
chat = ChatOpenAI()


# Initialize an empty list to store classifications
classifications = []

for review in reviews:
    print('Classifying review: ' + review['review'])

    # Classify each review individually using the model
    response = chat.invoke([
        SystemMessage(content='''You are responsible for the classification of movie reviews.
                                 Please classify the following review as either
                                 "positive" or "negative". You must only respond
                                 with one word: either "positive" or "negative".'''),
        HumanMessage(content=review['review'])
    ])

    # Strip spaces and lowercase the response to handle potential formatting variations
    response_text = response.content.strip().lower()

    if response_text not in ['positive', 'negative']:
        # Throw an error if the classification is invalid
        raise Exception('Invalid classification: ' + response_text)

    # Append the classification to the list
    classifications.append(response_text)

print("Classifications: ", classifications)
```
